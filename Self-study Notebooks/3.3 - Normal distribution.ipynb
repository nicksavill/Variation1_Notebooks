A histogram of a set of data allows us to visualise the shape of the data. We can see roughly where the centre of the data lies and how spread out the data are. We can see the number of peaks (also known as modes) and we can see if the data are skewed to the left or right. Finally a histogram allows us to see if there are any outliers, values that are unusually far from the mean.

Some of the datasets you've looked at in the self-study and exercise notebooks and in the workshop have a bell shape when plotted in a histogram. In other words, they have a single, central peak with most values clustering close to the mean. As we move away from the mean on either side there are fewer and fewer values.

Data that exhibit this bell-shaped histogram have a special name: They are said to be **normally distributed**. The histogram of such data also has a special name: A **Normal distribution**. 

Normally distributed data are ubiquitous in the natural world. The Normal distribution is one of the most important distributions in Biology and statistics so you need to know about it and understand its properties.

Don't confuse the name "Normal" with its English connotations of being average or that other distributions are abnormal. It's just a name. The Normal distribution has other commonly used names such as "Bell curve" and "Gaussian distribution". It's the same thing, just different names.

## Read and plot the data in a histogram

We're going to read in the student height data in order to understand the properties of the Normal distribution.

The file `Datasets/college students.csv` contains the heights in centimetres of the sample of 125 female students. As always, the first thing to do is to see what it looks like in a histogram.

<div class="alert alert-info">

Read then run the code cell below to plot a histogram of the students' heights.
</div>

````python
import pandas as pd
import seaborn as sns

students = pd.read_csv('Datasets/college students.csv') # Read in the students' heights

g = sns.displot(students['height']) # Plot a histogram

# Add some annotation
g.ax.set_xlabel('Height (cm)')
g.ax.set_ylabel('Number of students')
g.ax.set_title('Frequancy distribution of heights of\n125 US female college students');
````

If you compare the seaborn histogram you just plotted to the living histogram you'll see that they don't quite match. ## Find the sample size, mean and standard deviation

Having plotted the data to see what it looks like the next thing to do is get the sample size (the number of students in the sample) and calculate the sample mean and standard deviation.

Remember, the sample size is usually written as *n*, the sample mean is usually written as $\bar{x}$ (pronounced x-bar), and the sample standard deviation is usually written as $s$. 

<div class="alert alert-info">

Run the code cell below to calculate the sample size, mean and sample standard deviation.
    
</div>

````python
n    = students['height'].count()   # The sample size
xbar = students['height'].mean()    # The sample mean
s    = students['height'].std()     # The sample standard deviation

print(f'sample size is {n} students')
print(f'mean height is {xbar:.2f} cm')
print(f'st. dev. is {s:.2f} cm')
````

Notice we've printed the sample mean and standard deviation to 2dp as the data were measured to the nearest tenth of a centimetre.

So the average height of this sample of 125 female students is 164.19 cm. If you look at where this value lies in the living and seaborn histograms you'll see that it roughly corresponds to the peak in the distribution.

The standard deviation is 6.55 cm. But what does a value of 6.55 cm actually mean?

## What does the standard deviation represent? The 68-95-99.7% rule <a name="rule"></a>

We know what standard deviation is: It's a measure of how spread out, or variable, the data are - the more spread out the data the larger the standard deviation. But what does a standard deviation of 6.55 cm in student heights actually represent? 

The answer is the 68-95-99.7% rule.

The following figure displays this rule graphically. Look at it carefully, particularly the tick labels on the $x$-axis. Remember $s$ is the usual symbol for standard deviation.



The rule says that if data are normally distributed then
* 68% of the data lie within one standard deviation of the mean.
* 95% of the data lie within two standard deviations of the mean.
* 99.7% of the data lie within three standard deviation of the mean.

This means that all of the data (100% of it) are plotted in the histogram. This sounds obvious, but it's very useful to remember as we'll see in the Exercises. 

---

The 68-95-99.7% rule can sounded rather confusing. So let's think about what it means for the female college students. 

Let's start with the 68% part of the rule:
* 68% of the data lie within one standard deviation of the mean.

<div class="alert alert-info">

**"68% of the data"** means 68% of 125 students which is (run the following code)
</div>

````python
0.68 * n
````

68% of 125 students is exactly 85 students.

**"Within one standard deviation of the mean"** is the region from the mean minus one standard deviation to the mean plus one standard deviation.  

The mean minus one standard deviation is $\bar{x}-s$.

The mean plus one standard deviation is $\bar{x}+s$.

<div class="alert alert-info">

Let's calculate these in the following code cell.
</div>

````python
print( f'mean minus one st. dev. = {xbar - s:.2f} cm' )
print( f'mean plus one st. dev. = {xbar + s:.2f} cm' )
````

For the students the mean height is $\bar{x}$ = 164.19 cm and the standard deviation is $s$ = 6.55 cm (calculated above) so

* mean minus one standard deviation is 164.19 - 6.55 which equals 157.64 cm
* mean plus one standard deviation is 164.19 + 6.55 which equals 170.74 cm

---

Putting these two bits of information together we can say that when  female student heights are normally distributed then the heights of 85 of them will lie between 157.64 cm and 170.74 cm. That's the students in the central, unshaded part of the living histogram.

<div>

</div>

Let's check if the students' heights actually obey the rule.

Which means we need to find the frequency of student heights between 157.64 cm and 170.74 cm. If the students' heights are normally distributed then we would expect about 85 of them to have heights within this range. We say "about" because it doesn't have to be exactly 85 students. We're working with real messy data and nothing is ever exact.

How do we do that?

We could count the number students in the central region by eye. But that's tedious, and we have a computer to help us.  

We saw in the [Counting values notebook](2.4%20-%20Counting%20values.ipynb#counting-values-within-a-range) how to count the number of values in a dataset that lie within a certain range. Let's reuse and modify that code to count the number of students between 157.64 cm and 170.74 cm.

<div class="alert alert-info">

1. Look at the code below to try and understand what it does. 
2. Run the code.
</div>

````python
# Set tally of students within one standard deviation of the mean to zero.

count = 0 

# Loop through student heights one at a time.

for height in students['height']:
    
    # If this student's height is within one standard deviation of the mean increment tally by 1. 

    if xbar - s < height < xbar + s:
        count += 1

print( f'There are {n} students in total' )
print( f'{count} students have heights between {xbar - s:.2f} cm and {xbar + s:.2f} cm' )
print( f'Therefore {count / n:.1%} of students have heights within one standard deviation of the mean' )
````

So, 87 students (or 69.6% of them) have heights within one standard deviation of the mean. That's pretty close to 85. So we have some evidence that their heights are actually normally distributed.

## Aside: Try not to hard-code statistics and data

Notice in the code we wrote:

```python
    if xbar - s < height < xbar + s:
```

That is, we used the variables `xbar` and `s`, which we had calculated beforehand, to test if `height` was within one standard deviation of the mean. 

Instead, we could have written:

```python
    if 157.64 < height < 170.74:
```

This is called **hard-coding**, that is, embedding statistics and data into your code. This is **not good coding practice**. It makes the code specific to this particular dataset. If you want to apply your code to another dataset on heights, or you wanted to remove outliers, say, you first need to calculate these numbers then write them into your code. Using `xbar` and `s` instead means your code is reuseable. This is called **soft-coding** and is good practice.

## Estimating a standard deviation by eye

Estimating things by eye is an extremely useful skill to learn. If a dataset is roughly normally distributed we can estimate its mean and standard deviation by eye. 

Estimating the mean from a histogram is straightforward. The mean is the centre of the data; roughly where the peak occurs. 



Eye-balling the histogram we see the mean is approximately 165 cm (the actual mean is 164.19 cm).

Estimating the standard deviation is a little harder. We know that almost all (95%) of the data lie within two standard deviations of the mean. That's a range of 4 standard deviations from $-2s$ to $+2s$. We can therefore get a rough estimate of the standard deviation by subtracting the minimum value from the maximum value and dividing by 4, like so:

$$s \approx \frac{\mathrm{max} - \mathrm{min}}{4}$$

The squiggly equals sign means "approximately equal to", i.e., *s* is approximately equal to the maximum value minus the minimum value divided by 4. Eye-balling the histogram we see that the maximum value is about 183 cm and the minimum value is about 150 cm. Substituting these into the above formula gives

$$s \approx \frac{183 - 150}{4} = 8.25\ \mathrm{cm}$$

The actual standard deviation is 6.5 cm. So not far off.
